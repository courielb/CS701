```python
gtd['Counts_Weapon'] = gtd.groupby(['weaptype1_txt'])['related'].transform('count')
pivot1 = pd.pivot_table(gtd,index=["weaptype1_txt"],values=["Counts_Weapon"])
ax = pivot1[['Counts_Weapon']].plot(kind='bar', title ="Frequency of Weapon Type", figsize=(7, 7), legend=False, fontsize=12)
ax.set_xlabel("Weapon Type", fontsize=12)
ax.set_ylabel("Count of Attacks", fontsize=12)
plt.show()
pivot1
```

```python
gtd['Counts_Region'] = gtd.groupby(['region_txt'])['related'].transform('count')
pivot2 = pd.pivot_table(gtd,index=["region_txt"],values=["Counts_Region"])
ax = pivot2[['Counts_Region']].plot(kind='bar', title ="Frequency of Regional Attacks", figsize=(7, 7), legend=False, fontsize=12)
ax.set_xlabel("Region", fontsize=12)
ax.set_ylabel("Count of Attacks", fontsize=12)
plt.show()
pivot2
```

```python
gtd['Counts_AttackType'] = gtd.groupby(['attacktype1_txt'])['related'].transform('count')
pivot3 = pd.pivot_table(gtd,index=["attacktype1_txt"],values=["Counts_AttackType"])
ax = pivot3[['Counts_AttackType']].plot(kind='bar', title ="Attack Type", figsize=(7, 7), legend=False, fontsize=12)
ax.set_xlabel("Attack", fontsize=12)
ax.set_ylabel("Count of Attack Types", fontsize=12)
plt.show()
pivot3
```

```python
gtd['Counts_TargetType'] = gtd.groupby(['targtype1_txt'])['related'].transform('count')
pivot4 = pd.pivot_table(gtd,index=["targtype1_txt"],values=["Counts_TargetType"])
ax = pivot4[['Counts_TargetType']].plot(kind='bar', title ="Target Type", figsize=(7, 7), legend=False, fontsize=12)
ax.set_xlabel("Target", fontsize=12)
ax.set_ylabel("Count of Target Types", fontsize=12)
plt.show()
pivot4
```
